# Aus welchem Jahr stammen die meisten Publikationen?

```python
import pandas as pd
```

```python
title = pd.read_csv("titles.csv") # Einlesen der CSV Datei
```

```python
title
```

```python
date = title['Date of creation/publication'] 
# Komplette Tabelle wird auf die relevante Spalte mit den Publikationsjahren reduziert
```

```python
counts = pd.Series(date).value_counts().nlargest(5) #Zählt wie oft die Publikationsjahre vorkommen (mit der Funktion Value_counts)  
# und zeigt die Top 5 der Jahre mit den meisten Publikationen (mit nlargest)
print(counts) 
# Lösung der Frage 1
```
