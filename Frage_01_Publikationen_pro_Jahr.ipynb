# Aus welchem Jahr stammen die meisten Publikationen?
---

```python
import pandas as pd
```

```python
title = pd.read_csv("data/titles.csv")
# Einlesen der CSV Datei
```

```python
title
```

```python
date = title['Date of creation/publication'] 
# Mit der Ausgabe von title wird uns die komplette Tabelle angezeigt
# Allerdings benötigen wir für die Beantwortung der Frage lediglich die Spalte mit den Publikationsjahren
# Deswegen schränken wir die Tabelle nun so ein, dass uns nur noch die relevanten Daten ausgegeben werden
```

```python
counts = pd.Series(date).value_counts().nlargest(5)
# Die Funktion "Value_counts" zählt, wie häufig die jeweiligen Publikationsjahre vorkommen
# Zur besseren Übersicht haben wir uns dafür entschieden, dass uns nicht die Häufigkeit aller Publikationsjahre angezeigt werden soll, sondern stattdessen die Top 5
# Um die Top 5 der Jahren mit den meisten Publikationen herauszufinden, nutzen wir die Funktion "nlargest"
# Die Zahl in der Klammer bestimmt dabei, wie viele Werte uns angezeigt werden soll (in unserem Fall wie bereits erwähnt die Top 5) 
print(counts) 
# Lösung der Frage 1
```

---
# Grafische Darstellung der Lösung

```python
counts.plot.barh()
# Für die grafische Darstellung der Top 5 haben wir uns für die Bar Chart (Säulendiagramm) entschieden
# Wir haben uns dabei für eine horizontale Bar Chart, d. h. mit den Jahren an der Y-Achse, entschieden
# Diese ist unserer Meinung nach besser zu lesen und lässt die Unterschiede in den verschiedenen Jahren deutlicher erkennen
```
